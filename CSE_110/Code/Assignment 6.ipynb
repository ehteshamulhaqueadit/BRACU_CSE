```python
#Task 8
dict1=input()     
dict1=dict1[1:-1]   
dict1=dict1.split(",")
x={}
for i in dict1:
  i=i.split(":")
  k=i[0]    
  v=i[1]
  k=k.strip()
  k=k[1:-1]
  v=v.strip()
  x[k]=int(v)
y=0
j=0
for i in x.values():
  y+=i
  j+=1
print("Average is",y//j)
```
