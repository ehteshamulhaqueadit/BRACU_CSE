```python
#Task 6
x=input("")
x=x[1:-1]
x=[i for i in x.split(", ")]
y=[]
for i in x:
  i=int(i)
  y.append(i)
print("My list: ",y)
count1=0
count2=-1
l1=y[count1]
while count1<len(y):
  if y[count1]>=l1:
    l1=y[count1]
    count2+=1
  count1+=1
print("Largest number in the list is ",l1," which was found at index ",count2)
```

```python
#Task 10
x=[i for i in input("").split(", ")]
y=[]
z=[]
for i in x:
  i=int(i)
  y.append(i)
print("Input list: ",y)
pr=y[0]
for i in y:
  if i!=pr:
    z.append(pr)
    pr=i
print("Modified list: ",z)
```
