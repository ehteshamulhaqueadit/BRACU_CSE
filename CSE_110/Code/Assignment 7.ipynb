```python
#Task 1
def even_checker(user_input):
  if user_input%2==0:
    print("Even!!")
  else:
    print("Odd!!")
even_checker(int(input()))
```

```python
#Task 2
def fibonacci(n):
  stop=0
  finish=1
  count=0
  print(0,end=" ")
  print(1,end=" ")
  while True:
    start=stop
    stop=finish
    finish=start+stop
    if finish<=n:
      print(finish,end=" ")
    else:
      break
    if count==0:
      finish=1
      count+=1
fibonacci(int(input()))
```

```python
#Task 3
def foo_moo(n):
  if n%2==0 and n%3==0:
    return "FooMoo"
  if n%2==0:
    return "Foo"
  if n%3==0:
    return "Moo"
  else:
    return "Boo"
print(foo_moo(int(input())))
```

```python
#Task 4
def function_name(user):
  upl=0
  lpl=0
  for i in user:
    i=ord(i)
    if i>=65 and i<=90:
      upl+=1
    if i>=97 and i<=122:
      lpl+=1
  print("No. of Uppercase characters :",upl)
  print("No. of Lowercase characters : ",lpl)
function_name(input())
```

```python
#Task 5
def calculate_tax(age,salary,job):
  if age<18:
    return 0
  elif job.lower()=="president":
    return 0
  elif salary<10000:
    return 0
  elif 20000>salary>10000:
    tax=(salary/100)*5
    return tax
  else:
    tax=(salary/100)*10
    return  tax
  
print(calculate_tax(int(input()),int(input()),input()))
```

```python
#Task 6
def function_name(days):
  for i in ["years,","months and","days"]:
    if i=="years,":
      print(int(days/365),i,end=" ")
    elif i=="months and":
      print(int((days%365)/30),i,end=" ")
    else:
      print(int((days%365)%30),"days")
function_name(int(input()))
```

```python
#Task 7
def show_palindrome(a):
  x=""
  for i in range(1,a):
    x+=str(i)
  for i in range(a,0,-1):
    x+=str(i)
  return x
print(show_palindrome(int(input())))
```

```python
#Task 8
def show_palindromic_triangle(n):
  
  for j in range(1,n+1):
    x=""
    for i in range(n+1-j,-1,-1):
      x+="    "
    for i in range(1,j):
      x=x+str(i)+"   "
    for i in range(j,0,-1):
      x=x+str(i)+"   "
    print(x)
show_palindromic_triangle(int(input()))
```

```python
#Task 9 
def area_circumference_generator(r):
  import math
  area=math.pi * (r**2)
  circumference=2* math.pi * r
  final=(area,circumference)
  return final
final=area_circumference_generator(int(input()))
print(final)
area,circumference=final
print("Area of the circle is ",
      area," and circumference is ",circumference)
```

```python
#Task 10
def make_square(p_tuple):
    p1,p2=p_tuple
    dict1={}
    for i in range(p1,p2+1):
        dict1[i]=i**2
    return dict1
print(make_square((5,9)))
```

```python
#Task 11
def rem_duplicate(p_tuple):
    l1=[]
    for i in p_tuple:
        if i not in l1:
            l1.append(i)
    return tuple(l1)
print(rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a',
                     'a', [1,2])))
```

```python
#Task 12
def function_name(p_list):
    l1=[]
    remove_count=0
    for i in p_list:
        count=l1.count(i)
        if count<2:
            l1.append(i)
        else:
            remove_count+=1
    print("Removed:",remove_count)
    return l1
print(function_name([1, 2, 3, 3, 3, 3, 4, 5, 8, 8]))
```

```python
#Task 13
def function_name(p1,p2,p3):
    if '"' in p1 or "'" in p1:
        p1=p1[1:-1]
    if p1=="+":
        return p2+p3
    elif p1=="-":
        if p3>p2:
            return p3-p2
        if p2>p3:
            return p2-p3
    elif p1=="/":
        if p3>p2:
            return p3/p2
        if p2>p3:
            return p2/p3
    elif p1=="*":
        return p2*p3
print(function_name(input(),int(input()),int(input())))
```

```python
#Task 14
def function_name(p1,p2):
    count,final_string,p1,removed=0,p1[1],p1[2:-1],""
    for i in p1:
        count+=1
        if count%p2!=0:
            final_string+=i
        else:
            removed+=i
    return final_string+removed
print(function_name(input(),int(input())))
```

```python
#Task 15
def function_name(order_items,location="Dhanmondi"):
    dict1={"Rice":105,
     "Potato":20,
     "Chicken":250,
     "Beef":510,
     "Oil":85}
    sum=0
    for i in order_items:
        sum+=dict1[i]
    if location=="Dhanmondi":
        sum+=30
    else:
        sum+=70
    return sum
x=[i for i in input("")[1:-1].split(", ")]
y=[]
for i in x:
    i=i[1:-1]
    y.append(i)
print(function_name(y))
```
