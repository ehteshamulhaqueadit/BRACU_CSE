```python
#d using while loop (incompleat)
count=18
while count<=63:
    if count==63 and (count%2)!=0:
        print(count*(-1),end=(""))
    elif count!=63 and count%2==0:
        print(count,end=(","))
    else:
        print(count*(-1),end=(","))
    count=count+9
```

Task 14

```python
number=int(input("Please Enter the number"))
sum=0
for i in range(1,number+1):
    if number%i==0 and i!=number:
        sum=sum+i
if sum==number:
    print(number,"is a perfect number")
elif sum!=number:
    print(number,"is not a perfect number")
```
