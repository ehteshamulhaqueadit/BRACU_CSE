```python
#Task 1 bubble sort

my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
for i in range(len(my_list)-1):
    for j in range(len(my_list)-1-i):
        if my_list[j]>my_list[j+1]:
            temp=my_list[j+1]
            my_list[j+1]=my_list[j]
            my_list[j]=temp
print(my_list)
```

```python
#Task 2 selection sort

my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
for i in range(0,len(my_list)-1,1):
    m=i
    for j in range(i,len(my_list)):
        if my_list[m]>my_list[j]:
            m=j
    temp=my_list[i]
    my_list[i]=my_list[m]
    my_list[m]=temp
print(my_list)
```

```python
#Task 4

def selection_sort(my_list):
    for i in range(0,len(my_list)-1,1):
        m=i
        for j in range(i,len(my_list),1):
            if my_list[m]>my_list[j]:
                m=j
        temp=my_list[i]
        my_list[i]=my_list[m]
        my_list[m]=temp
    return my_list

def odd_even(sitting_list):
    even=[]
    odd=[]
    count=0
    for i in sitting_list:
        if count%2==0:
            even.append(i)
        else:
            odd.append(i)
        count+=1
    return even,odd

def arrange(even,odd):
    temp=[]
    count=0
    if len(sitting_list)%2==0:
        for i in even:
            temp.append(i)
            temp.append(odd[count])
            count+=1
    else:
        for i in odd:
            temp.append(i)
            temp.append(even[count])
            count+=1
    return temp

sitting_list = [10,30,20,70,11,15,22,16,58,100,12,56,70,80]
even,odd=(odd_even(sitting_list))
even,odd=selection_sort(even),selection_sort(odd)
odd.reverse()

print(arrange(even,odd))
```

```python
#Task 5

def selection_sort(my_list,num_index):
    for i in range(0,len(my_list)-1,1):
        m=i
        for j in range(i,len(my_list),1):
            if my_list[m][num_index]<my_list[j][num_index]:
                m=j
        temp=my_list[i][num_index]
        my_list[i][num_index]=my_list[m][num_index]
        my_list[m][num_index]=temp

        temp=my_list[i][0]
        my_list[i][0]=my_list[m][0]
        my_list[m][0]=temp
    return my_list

def take_input():
    x=input().upper()
    if x=="CSE110":
        x=1
    elif x=="PHY111":
        x=2
    elif x=="MAT110":
        x=3
    return x

def print_name(final_list):
    for i in final_list:
        print(i[0])


lst = [ ["Alan", 95, 87, 91], ["Turing", 92, 90, 83], ["Elon", 87, 92, 80], ["Musk", 85, 94, 90] ]
num_index=take_input()
final_list=selection_sort(lst,num_index)
print_name(final_list)
```
